```python

first_name = input("name is")
last_name = input("last is")
score =  input("score is :")
score = int(score)
if(score >=80):
    print(first_name,last_name,"A")
elif(score >=75):
    print(first_name,last_name,"B+")
elif(score >=70):
    print(first_name,last_name,"B")
elif(score >=65):
    print(first_name,last_name,"C+")
elif(score >=60):
    print(first_name,last_name,"C")
elif(score >=55):
    print(first_name,last_name,"D+")
elif(score >=50):
    print(first_name,last_name,"D")
else:
    print(first_name,last_name,"F")    

    
```
